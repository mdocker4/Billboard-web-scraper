# 1. Import packages

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

# 2. Find the page

```python
page = requests.get("https://www.billboard.com/charts/hot-100")
soup= BeautifulSoup(page.content, 'html.parser')
```

# 3. Parse the HTML for the chart container

```python
chart=soup.find("div", {"class":"chart-list container"})
```

```python
print(chart)
```

# 4. Requirements for dataframe

#### chart spot this week
#### song name
#### artist name
#### last week place
#### peak place
#### weeks on chart

# 5. Chart Element list

```python
items=chart.find_all(class_='chart-list__element display--flex')
```

# 6. Testing the 1st element

```python
print(items[0])
```

# 7. Clean the text

```python
print(items[0].find(class_="chart-element__rank__number").get_text())

print(items[0].find(class_="chart-element__information__song text--truncate color--primary").get_text())

print(items[0].find(class_="chart-element__information__artist text--truncate color--secondary").get_text())

print(items[0].find(class_="chart-element__meta text--center color--secondary text--last").get_text())

print(items[0].find(class_="chart-element__meta text--center color--secondary text--peak").get_text())

print(items[0].find(class_="chart-element__meta text--center color--secondary text--week").get_text())
```

# 8. List comprehension for every column

```python
rank_numbers=[i.find(class_="chart-element__rank__number").get_text() for i in items]
print(rank_numbers)
```

```python
song_names=[i.find(class_="chart-element__information__song text--truncate color--primary").get_text() for i in items]
print(song_names)
```

```python
artist_names=[i.find(class_="chart-element__information__artist text--truncate color--secondary").get_text() for i in items]
print(artist_names)
```

```python
last_week_place=[i.find(class_="chart-element__meta text--center color--secondary text--last").get_text() for i in items]
print(last_week_place)
```

```python
peak_place=[i.find(class_="chart-element__meta text--center color--secondary text--peak").get_text() for i in items]
print(peak_place)
```

```python
weeks_on_chart=[i.find(class_="chart-element__meta text--center color--secondary text--week").get_text() for i in items]
print(weeks_on_chart)
```

# 9. Turn the lists into a pandas dataframe

```python
Billboard_100=pd.DataFrame({
    
    "Rank":rank_numbers,
    "Song Name":song_names,
    "Artist Name": artist_names,
    "Last Week Place":last_week_place,
    "Peak Place":peak_place,
    "Weeks on Chart":weeks_on_chart
    
})
Billboard_100.to_csv('Billboard March 12')
```
